```python
from google.colab import files
import pandas

uploaded = files.upload()
uploaded


import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['total_words3.csv']))
df.head()
```

```python
#valores X
X = df.iloc[:, 2:7]
print(X)
#valores Y
Y = df.iloc[:, 7]
#print(Y)

# Codificar datos categ√≥ricos(Cambiamos strings por numeros)
from sklearn.preprocessing import LabelEncoder
#Codificar Sentimiento
labelencoder_X_country = LabelEncoder()
Y[:] = labelencoder_X_country.fit_transform(Y[:])
print(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20)
#print(y_train)
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
#Escalado de caracteristicas

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
print(X_train)
X_test = scaler.transform(X_test)
```

```python
# Build the model
from sklearn.naive_bayes import MultinomialNB

# Train the model
naive_bayes = MultinomialNB()
naive_bayes_fit = naive_bayes.fit(X_train, y_train)

from sklearn.metrics import confusion_matrix, balanced_accuracy_score

# Make predictions
train_predict = naive_bayes_fit.predict(X_train)
test_predict = naive_bayes_fit.predict(X_test)

def get_scores(y_real, predict):
  ba_train = balanced_accuracy_score(y_real, predict)
  cm_train = confusion_matrix(y_real, predict)

  return ba_train, cm_train 

def print_scores(scores):
  return f"Presicion: {scores[0]}\nMatriz de Confusion:\n {scores[1]}"

train_scores = get_scores(y_train, train_predict)
test_scores = get_scores(y_test, test_predict)


print(" Puntaje Entrenamiento")
print(print_scores(train_scores))
print("\n\n Puntaje Prueba")
print(print_scores(test_scores))
```
