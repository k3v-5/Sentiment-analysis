```python
# IMPORTS
import numpy as np
import random
import time
```

```python
## Declaraciones
tweets = []
for i in range(100):
    tweets.append([i, random.randint(30, 100)])


nproc = 2
l = 100
```

```python
#Class cromosoma for create population
class cromosoma:
  def __init__(self, n_genes):
    #Attributes
    self.n_genes = n_genes
    self.genes = []
    self.fitness = 0

    #Fill genes
    self.__fill_genes()
  
  #Methods
  def __fill_genes(self):
    for i in range(self.n_genes):
      self.genes += str(random.randint(0, 1))

  def get_genes(self):
    return self.genes
  
  def print_genes(self):
    print(self.genes)
```

```python
class population:
  def __init__(self, n_cromosomas, n_genes, tweets):
    #Attributes
    self.tweets = tweets
    self.n_cromosomas = n_cromosomas
    self.cromosomas = []
    self.n_mejores = 0

    self.__create_population(n_genes)

  #Methods
  def __create_population(self, n_genes):
    for i in range(self.n_cromosomas):
      self.cromosomas.append(cromosoma(n_genes))
      self.cromosomas[i].genes = self.tweets;
      self.tweets = self.tweets[1:] + self.tweets[:1]

  def print_population(self):
    for i in range(self.n_cromosomas):
      self.cromosomas[i].print_genes()
```
